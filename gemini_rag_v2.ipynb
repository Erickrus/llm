## Define Embedding Utils

```python
#@title load price data
with open("price.txt", "r") as f:
  lines = f.read().split("\n")

i = 0
prices, metadatas, ids = [], [], []
for line in lines:
  if line.strip() == "":
    continue
  prices.append(line)
  metadatas.append({"text": line})
  ids.append(str(i))
  i += 1
```

```python
#@title define chroma vector db client, init embedding model
# Reference: https://docs.trychroma.com/getting-started

import chromadb
from chromadb.utils import embedding_functions
chroma_client = chromadb.PersistentClient(path="kfc_vectordb")

# chromadb.errors.InvalidDimensionException: Embedding dimension 1024 does not match collection dimensionality
# https://github.com/PromtEngineer/localGPT/issues/541
from sentence_transformers import SentenceTransformer

embeddingModelName = 'jinaai/jina-embeddings-v2-base-zh' #@param ['BAAI/bge-large-zh-v1.5', 'jinaai/jina-embeddings-v2-base-zh']
#@markdown <img width=600px src='https://miro.medium.com/v2/resize:fit:4800/format:webp/1*ZP2RVejCZovF3FDCg-Bx3A.png' />
embeddingModel = SentenceTransformer('BAAI/bge-large-zh-v1.5')
```

```python
#@title define embedding_fn(documents)
#@markdown embedding fn usually takes a list of documents
def embedding_fn(documents):
  return [
    embeddingModel.encode(doc, normalize_embeddings=True).tolist()
    for doc in documents
  ]
collection = chroma_client.get_or_create_collection(name="kfc_collection")
```

```python
#@title process documents based on embeddings
# This is directly from embedding_fn
collection.add(
    documents=prices,
    embeddings= embedding_fn(prices),
    metadatas=metadatas,
    ids=ids
)
```

```python
#@title register Google API KEY
import google.generativeai as genai
import google.ai.generativelanguage as glm


genai.configure(api_key=GOOGLE_API_KEY)
gmodel = genai.GenerativeModel('gemini-pro')
```

```python
#@title define markdown
from bs4 import BeautifulSoup
from markdown import markdown

def markdown_to_text(md):
    try:
        soup = BeautifulSoup(markdown(md), features="html.parser")
        return soup.get_text()
    except:
        pass
    return md
```

## LLM RAG

```python
#@title define ask_price
def ask_price(text, products):
  #print(text)
  # check if text is related to anything in the embedding
  products = products.split()
  addedText = ""
  for product in products:
    print("检索:"+ product)

    results = collection.query(
        query_embeddings=embedding_fn([product]),
        n_results=2,
        include=['documents', 'distances', 'metadatas']
    )

    print("\n".join(results["documents"][0]))
    addedText += "\n".join(results["documents"][0]) + "\n"

  print()
  query = addedText + "\n请扮演一个客服，回答以下内容: "+ text
  print("问题重写为: "+query)
  response = gmodel.generate_content(query)
  answer = markdown_to_text(response.text)
  print()
  print("回答: "+ answer)
```

```python
#@title define ask_product
#@markdown

def ask_product(text):

  query = "请列出以下句子里有哪些产品，以回车分割: "+ text
  #print(query)
  response = gmodel.generate_content(query)
  answer = markdown_to_text(response.text)
  print("回答: "+ answer)
  return answer
```

```python
#@title Test
text = "我想买一份至珍全虾堡餐和一份老北京鸡肉卷，一共多少钱？" #@param {type:"string"}
print(text+"\n")
products = ask_product(text)

ask_price(text, products)
```
